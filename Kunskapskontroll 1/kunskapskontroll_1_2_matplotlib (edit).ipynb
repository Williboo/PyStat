### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

### Answer:

1st: Create figures and axes, and call methods on them "the object oriented" coding style.

2nd: Rely on pyplot to create and manage the figures and axes, then use pyplot functions. It's called "the pyplot style".

The object oriented way is generally recommended.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

### Answer:

Figure: Or a layout, it contains all the mentioned elements of a plot below. 

Axes: A specific area whitin a figure where data is plotted. 

Axis: A sort of framework for X, Y, Z axis like in a cordinate system.

Artist: Everything visible on the figure is an "artist". When the figure is rendered, all of the arists are drawn on the layout.

### When plotting in Matplotlib, what is the expected input data type?

### Answer:

Plotting functions expects "numpy.array" or "numpy.ma.masked_array" as input. Or objects that can be passed to "numpy.array".

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

### Using object oriented style:

```python
# Create x, and calculate y value
x = np.linspace(-4, 4, 200)
y = x**2

# Create figure and axes
fig, ax = plt.subplots()
# Plot data
ax.plot(x, y)

# Add label and title
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.set_title("y = x^2")

plt.show()
```

### Using pyplot style:

```python
# Create x, and calculate y value
x = np.linspace(-4, 4, 200)
y = x**2

# Create plot
plt.plot(x, y)

# Add label and title
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.title("y = x^2")

plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Create figure and subplots
fig, ax = plt.subplots(2, 1, figsize=(5, 10))

# Create scatter plot
ax[0].scatter(random_data_x, random_data_y)
ax[0].set_xlabel("x-axis")
ax[0].set_ylabel("y-axis")
ax[0].set_title("scatter plot")

# Create bar plot
ax[1].bar(names, values, color="blue")
ax[1].set_xlabel("fruit")
ax[1].set_ylabel("amount")
ax[1].set_title("Fruit bar plot")

plt.show()
```
